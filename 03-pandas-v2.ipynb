```python
#get a sample of the dataset

#in R:
# keep <- sample(ceiling(nrow(df)*.25), 1:nrow(df), FALSE)
# df[keep, ]

df.sample(frac=.25)
```

```python
#check total number of null values for each column

#in R: lapply(df, function(x) { length(which(is.na(x))) })

df.isnull().sum()
```

```python
#list only those with null values greater than zero

#in R: 
# null_values = lapply(df, function(x) { length(which(is.na(x))) })
# null_values[which(null_values > 0)]

null_values = df.isnull().sum()
null_values[null_values > 0]
```

```python
#rename column
#in R: names(df)[names(df) == "sofifa_id"] <- "FID"

df.rename(columns = {"sofifa_id": "FID"})
```

```python
#get median values, grouped by another column

#in R:
#library(dplyr)
# df %>%
#    group_by(nationality) %>%
#    summarize(median_age = median(age))

df.groupby("nationality")['age'].median()
```

```python
#get correlation between values

#in R: corr(df)

df.corr(method = 'pearson')
```

```python
#perform binning with speficic number of bins

#in R:
#library(dplyr)
#df %>% mutate(df = cut(age, breaks = 3))

pd.cut(df['age'], bins = 5).value_counts()
```

```python
#find largest values of a column 

#in R:
#library(dplyr)
#df %>% top_n(6, age)

df.nlargest(6, “Age”)
```
